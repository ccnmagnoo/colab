# Energy Proyects web-scrapping
source [seia](https://seia.sea.gob.cl/busqueda/buscarProyectoAction.php)
energy proyect web scrapping over

```python
import pandas as pd
import ssl
```

```python
data = []
URL = lambda page: f'https://seia.sea.gob.cl/busqueda/buscarProyectoAction.php?_paginador_fila_actual={page}'
def fetchData(page):
  fetch = pd.read_html(URL(page))
  return fetch[0]

for page in range(1,3):
  print.
  query = fetchData(page)
  data.append(query)
  print('tables loaded: ',len(data))
```
