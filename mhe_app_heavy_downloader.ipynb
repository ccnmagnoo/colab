```python
import pandas as pd
import requests
```

```python
##iter
YEAR_RANGE = range(2016,2022+1,1)
OPERATOR = 5
##query function returninf frame
def queryApi(year,operator):
  url:str = f'https://us-central1-myappt51.cloudfunctions.net/app/api/rooms?key=fa37Q3WqWOUoXCLXyAm0kRwmQaJ2&year={year}&op={operator}'
  res = requests.get(url)
  ##proceed return df
  query = res.json()
  df = pd.DataFrame(query["rooms"])
  return df
```

```python
##
global_data = pd.DataFrame()
##iter
for year in YEAR_RANGE:
  df = queryApi(year,OPERATOR)
  global_data = pd.concat([global_data,df])

global_data.sample(10)
```
